```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.isnull().sum())
```

```python
df['country'] = df['country'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')
```

```python
sns.countplot(data=df, x='type')
plt.title("Content Type Distribution")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(15, 5))
plt.title("Netflix Content by Release Year")
plt.show()
```

```python
from collections import Counter
genres = ','.join(df['listed_in'].dropna()).split(', ')
top_genres = Counter(genres).most_common(10)
pd.DataFrame(top_genres, columns=["Genre", "Count"]).plot(kind='bar', x='Genre', y='Count')
plt.title("Top 10 Genres")
plt.show()
```
